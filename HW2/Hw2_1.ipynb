```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Прочитайте дані за допомогою методу read_html з таблиці "Коефіцієнт народжуваності в регіонах України (1950—2019)"

url = 'https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8'

tables = pd.read_html(url, match='Коефіцієнт народжуваності в регіонах України', decimal=',', thousands='.')
table = tables[0]
table
```

```python
#Вивести перші рядки таблиці за допомогою методу head
table.head(5)
```

```python
#Визначте кількість рядків та стовпців у датафреймі (атрибут shape)
table.shape
```

```python
#Замініть у таблиці значення "—" на значення NaN
table.replace('—', np.nan, inplace=True)
table
```

```python
#Визначте типи всіх стовпців за допомогою dataframe.dtypes
table.dtypes
```

```python
#Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"
for t in table.columns:
    if table[t].isnull().any():
        table[t] = table[t].astype(float)
table.dtypes
```

```python
#Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)
for column in table.columns:
    print(f'{column}  {((table[column].isnull().sum()/ len(table[column])) * 100).round(2)}%')
```

```python
#Видаліть з таблиці дані по всій країні, останній рядок таблиці
table = table.drop(index=table.index[-1])
table
```

```python
#Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)
avg = table.mean(axis=0, numeric_only=True)
table = table.fillna(avg).round(1)
table
```

```python
#Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні
avg = table.mean(numeric_only=True)['2019']
table[['Регіон', '2019']][table['2019'] > avg]
```

```python
#У якому регіоні була найвища народжуваність у 2014 році?
table[["Регіон", "2014"]][table["2014"] == table["2014"].max()]
```

```python
#Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році
plt.title('Коефіцієнт народжуваності в Україні по регіонах у 2019 році')
table['2019'].plot(kind='bar', figsize=(10, 5), color='g')
plt.show()
```

```python
kherson = table[table['Регіон'] == 'Херсонська']
plt.plot(kherson.columns[1:], kherson.iloc[0,1:], 'o-', color = 'brown')
plt.title('Динаміка народжуваності у Херсонській області')
plt.xlabel('Рік')
plt.ylabel('Кількість народжених (на 1000 осіб)')
plt.show()
```

```python
fig, axs = plt.subplots()
axs.plot(table['Регіон'], table['2019'], label='2019', color = 'orange')
axs.plot(table['Регіон'], table['2000'], label='2000', color = 'y')
plt.plot(table['Регіон'], table['1990'], label='1990', color = 'g')
plt.plot(table['Регіон'], table['1950'], label='1950', color = 'r')
plt.xlabel('Регіон', fontsize='small')
plt.ylabel('Народжуваність', fontsize='small')
plt.title('Народжуваність по регіонах України')
plt.xticks(rotation=90)
plt.grid()
plt.legend()
plt.show()
```
